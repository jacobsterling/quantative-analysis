```python
import seaborn as sn
import yfinance as yf
```

```python
apple = yf.download("AAPL", start="2021-07-19", end="2024-08-19")

sn.histplot(apple['Close'], kde=True)
```

```python
tesla = yf.download("TSLA", start="2021-07-19", end="2024-08-19")

sn.histplot(tesla['Close'], kde=True)
```

```python
pct_apple = apple['Close'].pct_change()
pct_tesla = tesla['Close'].pct_change()
```

```python
sn.histplot(pct_apple, kde=True)
```

```python
sn.histplot(pct_tesla, kde=True)
```

```python
pct_apple.describe()
```

STD Apple = 1.72% == Volitility
Mean PCT = 0.074%

```python
pct_tesla.describe()
```

STD Tesla = 3.69% == Volitility
Mean PCT = 0.069%

```python
z_apple = (pct_apple - pct_apple.mean()) / pct_apple.mean()
z_tesla = (pct_tesla - pct_tesla.mean()) / pct_tesla.mean()
```

```python
from pandas import Series
from scipy.stats import norm

sd_apple = Series(norm.cdf(z_apple))
sd_tesla = Series(norm.cdf(z_tesla))
```
