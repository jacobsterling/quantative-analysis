```python
# TASK 3: Calculate the distances between a launch site to its proximities
from numpy import radians, sin, cos, arctan2, sqrt

def calculate_distance(lat1, lon1, lat2, lon2):
    # Radius of the Earth in kilometers
    R = 6371.0
    
    # Convert latitude and longitude to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Differences in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arctan2(sqrt(a), sqrt(1-a))
    
    # Calculate the distance
    distance = R * c
    
    return distance

# Define coordinates of proximities
proximities = {
    'Railway': (28.572646, -80.651225),
    'Highway': (28.573469, -80.646917),
    'Coastline': (28.573255, -80.646895)
}

# Calculate distances for each launch site
for index, site in spacex_df.iterrows():
    site_name = site['Launch Site']
    site_lat = site['Lat']
    site_lon = site['Long']
    
    for prox_name, prox_coords in proximities.items():
        distance = calculate_distance(site_lat, site_lon, prox_coords[0], prox_coords[1])
        print(f"Distance from {site_name} to {prox_name}: {distance:.2f} km")
    print()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
